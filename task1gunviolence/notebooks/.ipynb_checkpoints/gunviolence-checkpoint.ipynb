```python
import os
import pandas as pd
all_data = {}

for dirname, _, filenames in os.walk('../data'):
    for filename in filenames:
        data = pd.read_excel(os.path.join(dirname, filename))
        all_data[filename] = data

gun_data = all_data["gun.xlsx"]
pop_data = all_data["pop2018.xlsx"]
```

```python
if not gun_data.empty:
    print(gun_data.index)
else:
    raise EnvironmentError
```

```python
gun_data= gun_data.drop(columns='incident_id')
```

```python
sum_by_state = gun_data.groupby('state').sum()
```

```python
sum_by_state.sort_values('n_killed',ascending=False)
```

```python
sum_by_state['victims'] = sum_by_state['n_injured'] + sum_by_state['n_killed']
```

```python
sum_by_state.sort_values('victims',ascending=False)
```

```python
sum_by_state['murder_ratio'] = sum_by_state['n_killed'] / sum_by_state['victims']
```

```python
sum_by_state = sum_by_state.sort_values(by='n_killed',ascending= False)
```

```python
pop_data
```

```python
new_data = pd.merge(sum_by_state, pop_data, on=['state', ])
new_data = new_data.drop(columns=['n_injured', 'latitude', 'longitude', 'victims', ])
# new_data = new_data.set_index('state')
# new_data.to_csv("..\\data\\new_data.csv")
new_data = new_data.sort_values('murder_ratio', ascending=False )
new_data
```

```python
%matplotlib inline
import seaborn as sns
sns.set()
sns.lmplot('n_killed','population', data=new_data, fit_reg=False)
```
